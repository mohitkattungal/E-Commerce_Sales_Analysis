```python
import pandas as pd

df = pd.read_csv('Sample_Superstore.csv', encoding='latin1')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
```

```python
import matplotlib.pyplot as plt

df.groupby('Year')['Sales'].sum().plot()
plt.show()
```
